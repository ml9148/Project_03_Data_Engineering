```python
# Import dependencies
import pandas as pd
```

```python
# Load in the original data
df0 = pd.read_csv("Resources/Detail_listings.csv")
df0.head()
```

```python
# Drop columns not relevant to our database
df = df0.drop(columns=['listing_url','scrape_id','last_scraped','experiences_offered','thumbnail_url','medium_url','picture_url','xl_picture_url','host_url',
'host_acceptance_rate','host_thumbnail_url','host_picture_url','neighbourhood_group_cleansed','calendar_updated','has_availability','availability_30',
'availability_60','availability_90','availability_365','calendar_last_scraped','requires_license','license','jurisdiction_names'])
df.head()
```

```python
# Check primary key columns / NOT NULL col for null values
# listing id, host id, price 
print(df[df["id"].isna()])
print(df[df["host_id"].isna()])
print(df[df["price"].isna()])
```

```python
# Take a random sample to use for project
sample_df = df.sample(n=10000, random_state=42, ignore_index=True)
sample_df.head()
```

```python
# Check datatypes
for column in sample_df.columns:
    print(f"{column}: {sample_df[column].dtype}")
```

```python
# Convert datatypes


sample_df['host_name'] = sample_df['host_name'].astype('string')
sample_df['host_since'] = pd.to_datetime(sample_df['host_since'])
sample_df['host_location'] = sample_df['host_location'].astype('string')
sample_df['host_about'] = sample_df['host_about'].astype('string')
sample_df['host_response_time'] = sample_df['host_response_time'].astype('string')
sample_df['host_response_rate'] = sample_df['host_response_rate'].astype('string')
sample_df['host_is_superhost'] = sample_df['host_is_superhost'].astype('string')
sample_df['host_neighbourhood'] = sample_df['host_neighbourhood'].astype('string')
sample_df['host_listings_count'] = sample_df['host_listings_count'].astype('Int64')
sample_df['host_total_listings_count'] = sample_df['host_total_listings_count'].astype('Int64')
sample_df['host_verifications'] = sample_df['host_verifications'].astype('string')
sample_df['host_has_profile_pic'] = sample_df['host_has_profile_pic'].astype('string')
sample_df['host_identity_verified'] = sample_df['host_identity_verified'].astype('string')

```

```python
# Confirm data type changes
for column in sample_df.columns:
    print(f"{column}: {sample_df[column].dtype}")
```

```python
# Perform any further changes to data types / columns

# Host response rate column: 
# remove the percent symbol, trailing characters and whitespaces, and convert to integer
sample_df['host_response_rate'] = (sample_df['host_response_rate'].str.strip('%').str.strip('.0').str.strip()
                                   .replace('', pd.NA).astype(float).astype('Int64'))

# Superhost column:
# Convert t and f to boolean values and datatype
sample_df['host_is_superhost'] = sample_df['host_is_superhost'].fillna('f').map({'t': True, 'f': False})

# Host profile pic column:
sample_df['host_has_profile_pic'] = sample_df['host_has_profile_pic'].fillna('f').map({'t': True, 'f': False})

```

```python
# Confirm further data type changes
print(f"host_response_rate: {sample_df['host_response_rate'].dtype}")
print(f"host_is_superhost: {sample_df['host_is_superhost'].dtype}")
print(f"host_has_profile_pic: {sample_df['host_has_profile_pic'].dtype}")
```

```python
sample_df.to_csv("Resources/Listings_Cleaned_Sample.csv")
```
