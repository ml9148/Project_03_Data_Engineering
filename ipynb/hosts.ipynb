```python
# Import dependencies 
import pandas as pd
```

```python
# Load in the sample data
sample_df = pd.read_csv("Resources/Listings_Cleaned_Sample.csv")
sample_df.head()
```

```python
# Create DataFrame of just the Host-related columns
df_hosts = sample_df[['host_id','host_name','host_since','host_location',
                            'host_response_time','host_response_rate',
                            'host_is_superhost','host_neighbourhood',
                            'host_total_listings_count','host_verifications',
                            'host_has_profile_pic','host_identity_verified']].copy()
df_hosts.head()
```

```python
# Explore the data
print(df_hosts.shape)
df_hosts.dtypes
```

```python
# Clean columns / convert datatypes

# Host response rate column: 
df_hosts['host_response_rate'] = (df_hosts['host_response_rate'].astype('string').str.strip('%').str.strip()
                                   .astype(float))

# Superhost column:
# Convert t and f to boolean values and datatype
df_hosts['host_is_superhost'] = df_hosts['host_is_superhost'].fillna('f').map({'t': True, 'f': False})

# Host profile pic column:
df_hosts['host_has_profile_pic'] = df_hosts['host_has_profile_pic'].fillna('f').map({'t': True, 'f': False})

# Host identity verified column:
df_hosts['host_identity_verified'] = df_hosts['host_identity_verified'].fillna('f').map({'t': True, 'f': False})
```

```python
# Confirm data type changes
print(f"host_response_rate: {df_hosts['host_response_rate'].dtype}")
print(f"host_is_superhost: {df_hosts['host_is_superhost'].dtype}")
print(f"host_has_profile_pic: {df_hosts['host_has_profile_pic'].dtype}")
print(f"host_identity_verified: {df_hosts['host_identity_verified'].dtype}")
```

```python
# Confirm that some hosts have multiple properties
host_listings_in_sample = df_hosts['host_id'].value_counts()
host_listings_in_sample
```

```python
# Create a dictionary mapping host_id to its count
host_id_count_dict = host_listings_in_sample.to_dict()

# Map the host_id values to their counts and create a new column
df_hosts['host_listings_in_sample'] = df_hosts['host_id'].map(host_id_count_dict)
```

```python
# Reorder the new column 
df_hosts.insert(9, 'host_listings_in_sample', df_hosts.pop('host_listings_in_sample'))
```

```python
# Confirm the new column was created and re-located
df_hosts.head(3)
```

```python
# Drop the hosts with ids listed more than once and confirm
df_hosts.drop_duplicates(subset=['host_id'], inplace=True)
df_hosts['host_id'].value_counts().max()
```

```python
df_hosts.head(2)
```

```python
# Order the dataframe by host_id
df_hosts = df_hosts.sort_values(by='host_id')
```

```python
# Reset the index after cleaning
df_hosts.reset_index(drop=True, inplace=True)
```

```python
# Export the Hosts dataframe
df_hosts.to_csv('data/hosts.csv', index=False)
```
