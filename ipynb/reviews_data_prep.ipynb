```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/cdj3e/vu_bootcamp/Project_03_Data_Engineering/Resources/Data_Cleaning/Listings_Cleaned_Sample.csv")
df.head()
```

```python
df_clean=df.drop(columns=['Unnamed: 0','host_id','name','summary','space','description','neighborhood_overview','notes','interaction','house_rules','host_name','host_since','host_location','host_about','host_response_time','host_response_rate','host_is_superhost','host_listings_count','host_total_listings_count','host_verifications','host_has_profile_pic','host_identity_verified','street', 'neighbourhood','transit','access','host_neighbourhood','neighbourhood_cleansed','country','latitude','city','state','zipcode','market','smart_location','country_code','longitude','is_location_exact','property_type','room_type','accommodates','bathrooms','bedrooms','beds','bed_type','amenities','square_feet','price','weekly_price','monthly_price','security_deposit','cleaning_fee','guests_included','extra_people','minimum_nights','maximum_nights','instant_bookable','cancellation_policy','require_guest_profile_picture','require_guest_phone_verification','calculated_host_listings_count'])
df_clean.head()
```

```python
df_clean.drop_duplicates(subset=['id'], keep='first')
```

```python
df_clean.drop(df.loc[df['number_of_reviews']==0].index, inplace=True)
print(df_clean)
```

```python
for column in df_clean.columns:
    print(f'{column}: {df_clean[column].dtype}')
```

```python
df_clean['first_review'] = pd.to_datetime(df_clean['first_review'])
df_clean['last_review'] = pd.to_datetime(df_clean['last_review'])
df_clean['review_scores_rating'] = df_clean['review_scores_rating'].astype('Int64')
df_clean['review_scores_accuracy'] = df_clean['review_scores_accuracy'].astype('Int64')
df_clean['review_scores_cleanliness'] = df_clean['review_scores_cleanliness'].astype('Int64')
df_clean['review_scores_checkin'] = df_clean['review_scores_checkin'].astype('Int64')
df_clean['review_scores_communication'] = df_clean['review_scores_communication'].astype('Int64')
df_clean['review_scores_location'] = df_clean['review_scores_location'].astype('Int64')
df_clean['review_scores_value'] = df_clean['review_scores_value'].astype('Int64')
df_clean['reviews_per_month'] = df_clean['reviews_per_month'].astype('string')
```

```python
for column in df_clean.columns:
    print(f"{column}: {df_clean[column].dtype}")
```
